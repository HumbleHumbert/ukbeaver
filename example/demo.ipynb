```python
import polars as pl
import pandas as pd

# --- Polars read ---
df_polars = pl.read_csv(
    ph.supplement_table,
    separator="\t",
    ignore_errors=True,
    truncate_ragged_lines=True,
    infer_schema_length=0,
).select(["field_id", "value_type"])

# --- Pandas read ---
df_pandas = pd.read_csv(
    ph.supplement_table,
    sep="\t",
    usecols=["field_id", "value_type"],
    dtype=str  # avoid type mismatch issues
)

```

```python
# Convert Polars → pandas
df_polars_pd = df_polars.to_pandas()

# Fill NaN and cast everything to string for consistency
df_polars_pd = df_polars_pd.fillna("").astype(str)
df_pandas = df_pandas.fillna("").astype(str)
```

```python
import numpy as np

same_shape = df_polars_pd.shape == df_pandas.shape
same_content = df_polars_pd.equals(df_pandas)

print("Same shape:", same_shape)
print("Completely identical:", same_content)
```
